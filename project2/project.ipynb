```python
import numpy as np
import pandas as pd

# create a DataFrame from the given data
df = pd.read_excel('./data/car_stocks.xlsx')
df['Date'] = df['Date'].astype(str)
df['Manufactured Year'] = df['Date'].str[:4]
updated_car_stocks = df.groupby(['Manufacturer','Manufactured Year'])['Closing Price'].agg(['first','last']).reset_index()
updated_car_stocks.columns = ['Manufacturer','Manufactured Year','Starting Price','Ending Price']
updated_car_stocks['Stock Price Change'] = round((updated_car_stocks['Ending Price'] - updated_car_stocks['Starting Price']) * 100 / updated_car_stocks['Starting Price'])
updated_car_stocks['Stock Price Change'] = updated_car_stocks['Stock Price Change'].astype(str).str.rstrip('%').astype(float)
Median = updated_car_stocks.groupby('Manufactured Year')['Stock Price Change'].transform('median')
updated_car_stocks['Percentile'] = round(updated_car_stocks['Stock Price Change'] - Median)
updated_car_stocks['Percentile'] = updated_car_stocks['Percentile'].astype(int).astype(str) + '%'
updated_car_stocks['Stock Price Change'] = updated_car_stocks['Stock Price Change'].astype(int).astype(str) + '%'
updated_car_stocks
```
